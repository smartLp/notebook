
## **数学公式**
1. **行内公式**
> 我们使用 $E=mc^2$来计算能量
2. **行间公式**
> $$x = a_{1}^n + a_{2}^n + a_{3}^n$$

1. #### **上下标**
`^`表示上标，`_`表示下标，如果上下标的个数多余一个，可以用`{}`括起来作为一个整体
$$ a = b_{n+2} + 3$$
$$ c = n^{2c+3} +2a$$

2. ### **括号**
`()`,`[]`,`|` 都表示他们自己，由于`{}`有其特殊作用，一般大括号用`\lbrace\rbrace`表示
$$ f(x,y) = 100* \lbrace[(x+y)*3]-5\rbrace$$

3. ### **分数**

使用 `\frac{分子}{分母}`这样的语法
$$ f(x) = \frac{1.0 + 2x}{1.0 - 2x}$$

4. ### **开方**
开方使用 `\sqrt[次数]{被开方数}`
$$ \sqrt[3]{x^6} $$
$$ \sqrt{x^2+2} $$
